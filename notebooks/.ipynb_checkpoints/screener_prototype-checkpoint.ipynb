# CSP Screener Prototype

Simple notebook to test yfinance rate limiting and screener logic locally.

```python
import time
import yfinance as yf
from datetime import date
import pandas as pd

# Test symbols
SYMBOLS = ["ADBE", "AMZN", "AVGO", "META", "MSFT", "NVDA", "PLTR", "PYPL", "TSM", "UNH"]
```

## Test 1: Check yfinance response headers

Yahoo doesn't expose rate limit headers directly, but we can check for errors.

```python
def test_single_ticker(symbol: str):
    """Test a single ticker and return timing + error info."""
    result = {
        "symbol": symbol,
        "info_time": None,
        "options_time": None,
        "chain_times": [],
        "error": None,
        "info_keys": None,
    }
    
    try:
        # Test ticker.info
        t0 = time.monotonic()
        ticker = yf.Ticker(symbol)
        info = ticker.info
        result["info_time"] = time.monotonic() - t0
        
        if info:
            result["info_keys"] = list(info.keys())[:10]  # first 10 keys
            price = info.get("currentPrice") or info.get("regularMarketPrice")
            result["price"] = price
        else:
            result["error"] = "No info returned"
            return result
        
        # Test ticker.options
        t0 = time.monotonic()
        expirations = ticker.options
        result["options_time"] = time.monotonic() - t0
        
        if not expirations:
            result["error"] = "No options data"
            return result
        
        result["expirations"] = expirations[:3]
        
        # Test option_chain calls
        for exp in expirations[:2]:  # only test first 2 expirations
            t0 = time.monotonic()
            try:
                chain = ticker.option_chain(exp)
                result["chain_times"].append({
                    "expiry": exp,
                    "time": time.monotonic() - t0,
                    "puts_count": len(chain.puts) if chain.puts is not None else 0,
                })
            except Exception as e:
                result["chain_times"].append({
                    "expiry": exp,
                    "error": str(e),
                })
            time.sleep(1)  # small delay between chains
            
    except Exception as e:
        result["error"] = str(e)
    
    return result

# Test one symbol
result = test_single_ticker("AAPL")
result
```

## Test 2: Rate limit detection

Run multiple tickers with timing to see when rate limits kick in.

```python
def test_rate_limits(symbols: list[str], delay_between: float = 2.0):
    """Test multiple symbols and track when failures occur."""
    results = []
    
    for i, symbol in enumerate(symbols):
        print(f"[{i+1}/{len(symbols)}] Testing {symbol}...")
        
        if i > 0:
            time.sleep(delay_between)
        
        result = test_single_ticker(symbol)
        results.append(result)
        
        if result["error"]:
            print(f"  ERROR: {result['error']}")
        else:
            print(f"  OK: price={result.get('price')}, info_time={result['info_time']:.2f}s")
    
    return results

# Run test with first 5 symbols
results = test_rate_limits(SYMBOLS[:5], delay_between=3.0)

# Summary
print("\n=== SUMMARY ===")
for r in results:
    status = "ERROR" if r["error"] else "OK"
    print(f"{r['symbol']}: {status}")
```

## Test 3: Simple CSP Screener

```python
def scan_for_csp(
    symbol: str,
    min_dte: int = 30,
    max_dte: int = 90,
    min_otm_pct: float = 5.0,
    min_premium_pct: float = 0.5,
) -> list[dict]:
    """Scan for cash-secured put opportunities."""
    opportunities = []
    
    try:
        ticker = yf.Ticker(symbol)
        info = ticker.info
        
        if not info:
            print(f"No info for {symbol}")
            return []
        
        price = info.get("currentPrice") or info.get("regularMarketPrice")
        if not price:
            print(f"No price for {symbol}")
            return []
        
        print(f"{symbol}: ${price:.2f}")
        
        expirations = ticker.options
        if not expirations:
            print(f"  No options for {symbol}")
            return []
        
        today = date.today()
        
        for exp_str in expirations:
            exp_date = date.fromisoformat(exp_str)
            dte = (exp_date - today).days
            
            if not (min_dte <= dte <= max_dte):
                continue
            
            time.sleep(1.5)  # delay between chain calls
            
            try:
                chain = ticker.option_chain(exp_str)
            except Exception as e:
                print(f"  Chain error {exp_str}: {e}")
                continue
            
            if chain.puts is None or chain.puts.empty:
                continue
            
            for _, put in chain.puts.iterrows():
                strike = put.get("strike")
                bid = put.get("bid")
                iv = put.get("impliedVolatility")
                
                if bid is None or bid <= 0:
                    continue
                
                otm_pct = (price - strike) / price * 100
                
                if otm_pct < min_otm_pct:
                    continue
                
                premium_pct = (bid / strike) * 100
                
                if premium_pct < min_premium_pct:
                    continue
                
                ann_roc = (bid / strike) * (365 / dte) * 100
                
                opportunities.append({
                    "symbol": symbol,
                    "strike": strike,
                    "expiry": exp_str,
                    "dte": dte,
                    "bid": bid,
                    "price": price,
                    "otm_pct": round(otm_pct, 2),
                    "premium_pct": round(premium_pct, 2),
                    "ann_roc": round(ann_roc, 2),
                    "iv": round(iv, 4) if iv else None,
                })
        
    except Exception as e:
        print(f"Error scanning {symbol}: {e}")
    
    return opportunities

# Test on one symbol
opps = scan_for_csp("AAPL", min_dte=30, max_dte=60, min_otm_pct=3.0)
print(f"\nFound {len(opps)} opportunities")
if opps:
    pd.DataFrame(opps).head(10)
```

## Test 4: Scan watchlist with rate limit handling

```python
def scan_watchlist(
    symbols: list[str],
    delay_between_tickers: float = 5.0,
    max_retries: int = 3,
):
    """Scan watchlist with rate limit handling."""
    all_opps = []
    failed = []
    
    for i, symbol in enumerate(symbols):
        print(f"\n[{i+1}/{len(symbols)}] Scanning {symbol}...")
        
        if i > 0:
            print(f"  Waiting {delay_between_tickers}s...")
            time.sleep(delay_between_tickers)
        
        for attempt in range(max_retries):
            try:
                opps = scan_for_csp(symbol, min_dte=30, max_dte=60, min_otm_pct=3.0)
                all_opps.extend(opps)
                break
            except Exception as e:
                err_msg = str(e).lower()
                is_rate_limit = "rate" in err_msg or "429" in err_msg or "too many" in err_msg
                
                if is_rate_limit and attempt < max_retries - 1:
                    backoff = delay_between_tickers * (2 ** attempt)
                    print(f"  Rate limited, retry {attempt+1}/{max_retries} in {backoff}s")
                    time.sleep(backoff)
                else:
                    print(f"  FAILED: {e}")
                    failed.append(symbol)
                    break
    
    return all_opps, failed

# Run on first 3 symbols to test
all_opps, failed = scan_watchlist(SYMBOLS[:3], delay_between_tickers=5.0)

print(f"\n\n=== RESULTS ===")
print(f"Opportunities: {len(all_opps)}")
print(f"Failed: {failed}")

if all_opps:
    df = pd.DataFrame(all_opps)
    df.sort_values('ann_roc', ascending=False).head(20)
```
